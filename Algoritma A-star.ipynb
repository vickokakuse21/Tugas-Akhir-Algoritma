```python
class Point:
    def __init__(self, name):
        self.name = name
        self.neighbors = []
        self.parent = None
        self.H = 0
        self.G = 0
        self.F = 0
        
    def __repr__(self):
        return f"Point_{self.name}"

class AStarAlg:
    def __init__(self, start_point, end_point):
        self.start_point = start_point
        self.end_point = end_point
        self.open_list = []
        self.close_list = []
    
    def calc_H(self, node):
        return 0
    
    def get_min_F(self):
        min_node = self.open_list[0]
        for node in self.open_list[1:]:
            if node.F < min_node.F:
                min_node = node
        return min_node
    
    def solve(self):
        #inisialisasi parameter
        self.open_list = [self.start_point]
        search_flag = False
        #mulai pencarian
        while self.open_list and (not search_flag):
        # Dapatkan node dengan perkiraan biaya total terkecil untuk pencarian
            current_node = self.get_min_F()
            
            if current_node == self.end_point:
                search_flag = True
                break
            self.open_list.remove(current_node)
            self.close_list.append(current_node)
            
            # Cari node tetangga dari node saat ini
            for data in current_node.neighbors:
                node, dist = data 
                # Lewati node yang telah dicari
                if node in self.close_list:
                    continue
                # Proses node sudah ada di open_list
                elif node in self.open_list:  
                # Hitung dan perbarui informasi node
                    tmp_G = current_node.G + dist
                    if node.G < tmp_G:
                        continue
                    node.parent = current_node
                    node.G = tmp_G
                    node.F = node.G + node.H
                # Proses node yang belum ditambahkan ke open_list
                else:
                    node.parent = current_node
                    node.G = current_node.G + dist
                    node.H = self.calc_H(node)
                    node.F = node.G + node.H
                    self.open_list.append(node)
       # Akhir pencarian
        return search_flag
        
    def trace_path(self):
        # Mundur dari akhir
        path = [self.end_point]
        node = self.end_point
        # Mundur ke titik awal dan berhenti
        while node != self.start_point:
            node = node.parent
            path.append(node)
        return path
```
